<h1>Multivariate Linear regression</h1>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
mean=df['bedrooms'].mean()
mean
```

```python
mean=np.floor(mean)
mean
```

```python
df['bedrooms'].fillna(mean,inplace=True)
df
```

```python
plt.scatter(df['area'],
           df['price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area V/S price")
plt.grid()
plt.show()
```

```python
plt.scatter(df['bedrooms'],
           df['price'])
plt.xlabel("Bedrooms")
plt.ylabel("Price")
plt.title("Bedrooms V/S price")
plt.grid()
plt.show()
```

```python
plt.scatter(df['age'],
           df['price'])
plt.xlabel("age")
plt.ylabel("Price")
plt.title("Age V/S price")
plt.grid()
plt.show()
```

```python
model=LinearRegression()
X=df.drop(['price'],axis=1)
y=df['price']
model.fit(X,y)
```

```python
model.predict([[3000, 3, 40]])
```

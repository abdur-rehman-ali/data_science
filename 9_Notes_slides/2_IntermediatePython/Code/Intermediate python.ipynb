<h2>Chapter 1: Matplotlib</h2>

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter([1,2,3,4,5,6],[2,3,4,4,6,7])
plt.show()
```

```python
plt.scatter([1,2,3,4,5,6],[2,3,4,4,6,7])
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("x v/s y")
plt.yticks([0,1,2,3,4,5,6,7,8])
plt.grid()
plt.show()
```

```python
values = [0,0.6,1.4,1.6,2.2,2.5,2.6,3.2,3.5,3.9,4.2,6]
plt.hist(values,bins=5)
plt.show()
```

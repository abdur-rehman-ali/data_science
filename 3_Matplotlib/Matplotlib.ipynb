<h1>Matplotlib</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

<h3>Scatter plot</h3>

```python
x = np.arange(1,100)
y = np.arange(101,200)
plt.scatter(x,y,marker="+",c='g')
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Scatter plot")
plt.savefig("scatter.png")
```

<h3>Simple plot</h3>

```python
x= np.arange(1,10)
y=x*x
plt.plot(x,y,"r*",linestyle="dashed",linewidth=2)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Plot")
```

<h3>Subplots</h3>

```python
x= np.arange(1,10)
y=x*x
plt.subplot(2,2,1)
plt.plot(x,y,"r+")

plt.subplot(2,2,2)
plt.plot(x,y,"g-")

plt.subplot(2,2,3)
plt.plot(x,y,"bo")
```

<h3>Bar plot</h3>

```python
x= np.array([1,3,5,7])
y= np.array([5,6,7,8])

x1= np.array([2,4,6,8])
y1= np.array([15,61,17,18])

plt.bar(x1,y1,width=0.1)
plt.bar(x,y,width=0.1)
plt.xlabel("X caption")
plt.ylabel("Y caption")
plt.title("Bar plot")
plt.show()
```

<h3>Histogram</h3>

```python
#A diagram consisting of rectangles whose area is proportional to the frequency of a variable 
#and whose width is equal to the class interval.
x= np.array([1,2,3,4,5,1,1,1,1,12,32,43,54,65,76,87,87,43,123,13,2,54,34,65])
plt.hist(x,rwidth=0.1)
plt.title("Histogram")
```

<h3>Pie plot</h3>

```python
# Data to plot
labels = 'Python', 'C++', 'R', 'Java'
sizes = [200, 200, 200, 200]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.3, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.2f%%')

plt.axis('equal')
plt.show()
```
